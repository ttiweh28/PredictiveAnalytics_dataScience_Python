```python
import mysql.connector
from mysql.connector import Error

#the try and except commands are used to catch and handle exceptions
try:
    connection = mysql.connector.connect(host='localhost',
                                         database='orders_tab',
                                         user='root',
                                         password='')

    sql_select_Query = "select * from transactions"
    
    #create a cursor object to execute the SQL commands
    cursor = connection.cursor()
    cursor.execute(sql_select_Query)
    
    #fetchall used to obtain all rows in the transactions table
    inventory = cursor.fetchall()
    print("Total number of rows in Inventory is: ", cursor.rowcount)

    print("\nPrinting each Inventory record")
    
    #finally iterate through all the records obtained to print each one of them
    for row in inventory:
        print("Id = ", row[0], )
        print("Name = ", row[1])
        print("Price  = ", row[2])
        print("Purchase date  = ", row[3], "\n")

except Error as e:
    print("Error reading data from MySQL table", e)
finally:
    if (connection.is_connected()):
        
        #close the database connection
        connection.close()
        
        #close the cursor object
        cursor.close()
        print("MySQL connection is closed")
```
