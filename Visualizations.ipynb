```python
import random
def coin_trial():
    heads= 0 
    if random.random() <= 0.5:
        heads+=1
    return heads
```

```python
def simulate(n):
    trials=[]
    for i in range(n):
        trials.append(coin_trial())
    return(sum(trials)/n)
```

```python
simulate(1000000) 
```
